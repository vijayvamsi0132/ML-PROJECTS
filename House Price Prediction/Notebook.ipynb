```python
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
param_grid={
    "criterion": ["squared_error","poisson","friedman_mse",],
    "splitter": ["best","random"],
    "max_depth": [None,10,20,30,40,50],
    "min_samples_split":[2,5,10],
    "min_samples_leaf":[1,2,4]
}
```

```python
tree_model=DecisionTreeRegressor()
```

```python
grid_tree=GridSearchCV(estimator=tree_model,param_grid=param_grid)
```

```python
grid_tree.fit(x_train,y_train)
```

```python
grid_tree
```

```python
grid_tree.best_params_
```

```python
y_pred=grid_tree.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
acc=mean_squared_error(y_test,y_pred)
print(acc)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
pred=lr.predict(x_test)
```

```python
mean_squared_error(y_test,pred)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rfrmodel=RandomForestRegressor()
```

```python
rfrgrid={
    "max_depth":[5,10,15],
    "n_estimators":[2,3,4,5,6,7,8,9,10]
}

```

```python
rfrgrid=GridSearchCV(rfrmodel,rfrgrid)
```

```python
rfrgrid.fit(x_train,y_train)
```

```python
rfrgrid.best_params_
```

```python
rfrpred=rfrgrid.predict(x_test)
mean_absolute_error(y_test,rfrpred)
```
